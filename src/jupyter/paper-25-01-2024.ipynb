```python
import numpy as np
import matplotlib.pyplot as plt

def generate_points(num_epicenters, points_per_cluster, map_size):
    epicenters = np.random.rand(num_epicenters, 2) * map_size

    points = []
    for epicenter in epicenters:
        cluster = epicenter + np.random.randn(points_per_cluster, 2) * map_size / 10
        points.extend(cluster)
    print(points)
    return np.array(points)

def plot_points(points, map_size):
    plt.scatter(points[:, 0], points[:, 1], s=10, c='blue', marker='o')
    plt.xlim(0, map_size)
    plt.ylim(0, map_size)
    plt.title('2D Map with Clustered Points')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()

# Set parameters
num_epicenters = 2
points_per_cluster = 30
map_size = 100

# Generate and plot points
points = generate_points(num_epicenters, points_per_cluster, map_size)
plot_points(points, map_size)
```
