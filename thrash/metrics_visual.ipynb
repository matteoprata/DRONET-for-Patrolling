```python
from src.evaluation.metrics_main import __data_matrix_multiple_exps

data = __data_matrix_multiple_exps(setup01, indv.DRONES_SPEED)
print(data.shape)
```

```python
import numpy as np
# We get the max delay for each target
n1 = np.max(data, axis=0)
print(n1.shape)
```

```python
# Old approach in which we average both over the targets and runs, but especially the average over the targets
# can be misleading and hide very severe delays
n2 = np.average(n1, axis=(0, 2))
n2.shape
```

```python
print(n2)
```

```python
# New approach in which we get the max delay caused by any of the targets.
# We average this value for each of the seeds
n3 = np.max(n1, axis=2)
n3 = np.average(n3, axis=0)
n3.shape
```

```python
print(n3)
```
